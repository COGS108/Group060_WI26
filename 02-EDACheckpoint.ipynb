### Global number of streaming platforms carrying an IP
   1. Important metrics, what units they are in, and what they mean: 
      >This dataset was created using information from the TMDb API and looks at how easy it is to access different major entertainment franchises. Each row in the dataset represents one IP, such as Marvel, Star Wars, or Rilakkuma. The main metric used in this project is the number of streaming platforms that carry at least one movie or TV show from that IP. This value is a count, measured in number of platforms. The number of streaming platforms is used as a way to measure accessibility. If an IP appears on many platforms, it is easier for people to watch without needing a specific subscription. If it appears on only one or two platforms, it is more limited and harder to access. To calculate this, broad IPs were broken into individual movies and shows, and all streaming providers across those titles were combined.

   3. Major concerns with the dataset:

Because of this, the accessibility of some IPs may be underestimated, and the results should be viewed as general comparisons instead of exact measurements.
