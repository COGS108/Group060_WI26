## Research Question

Given the US’s deep integration with global popular culture—particularly Japanese IPs—does historical US IP performance in the US merchandize market still align with present-day consumer engagement, as measured by cumulative US merchandize revenue of the top ten bestselling US IPs and Japanese IPs and their search interest in the US from 2020-2025?


## Background and Prior Work

While the US is one of the major entertainment markets in the world, its merchandise sector depends heavily on imported IPs. The US anime merchandise market reached an annual revenue of several hundred million dollars by 2025, yet that represents only a small portion of the global anime merchandise market.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) Instead, a large amount of US consumer demand for merchandise is met through imports from Asian countries such as Japan.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Japan is very known for its sustaining IPs, which can be seen with franchises such as Pokémon and One Piece that continue to have major commericial success even through the decades since these series have been released. Japan‘s anime merchandise market alone generated billions of dollars in annual revenue by the mid-2020s, and is still expected to grow due to domestic and international demand.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Specifically, Japanese anime IPs also constitute a substantial portion of cross-border merchandise purchases, showing their global appeal, including the US.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

This raises questions about the strength of domestic US IPs, including major franchises such as Disney and Marvel. Many U.S. merchandise have shorter product lifecycles, being quickly outdated once media exposure declines. With the increasingly globalized U.S. popular culture, American consumers are continuously attracted to foreign media such as anime, gearing their attention away from domestic IPs. This raises the question of whether US domestic IPs still align with modern day US consumer interest, or if that interest has now shifted to Japanese IPs.

In 2025, the cultural differences between the marketing strategies by the United States vs the Japanese were explored.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) The findings exhibited that the US marketing strategies included an emphasis on bold self-expression and straight-forward, relatable advertising tactics. On the other hand, Japanese marketing tactics put more focus on storytelling, culture, family, and simplicity. They also focus on highlighting the craftsmanship and quality of their products. This may have some relation to the importance of brand loyalty in Japan compared to America. This emphasis on long-term relationships can be seen through the long-lasting dominance of brands such as Sony and Nintendo in Japanese society–and even amongst many Americans.

Research exploring the behavior of both American and Japanese children and their relationship with advertisement showed that there were many differences between the two audiences.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) In America, the children were more skeptical towards the advertisements and are more susceptible to direct messaging and reasoning. They also are more attracted to products that market towards the benefits of a child. In Japan, children are more vulnerable to advertisements and watch much less television. This means it is harder for them to recognize when something is an advertisement. Japanese children are more susceptible to emotional messaging that puts a larger emphasis towards the product benefitting their family’s needs.

Two years ago, the Japanese IP-driven Merchandising Market in only the United States was roughly valued at 19.2 Billion USD and has been projected to grow at a CAGR (Compound Annual Growth Rate) of 11.7 percent.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) The catalyst of this significant rise would be the increased popularization of anime in streaming platforms like Netflix, Crunchyroll and Hulu. Meanwhile, out of the top 10 global licensors of 2025, 8 out of 10 of the companies were American.<a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7) This includes the titan that is Disney–reaching 62 Billion USD globally.

## Hypothesis


The US IP's legacy in the US market does not align with present-day consumer engagement. With the increasing popularity of Japanese media in the United States, driven by modern distribution and marketing techniques such as streaming and social media–driven promotion, imported franchises such as Japanese anime IPs play an increasing role in merchandise sales and fan engagement. We expect the US merchandise revenue to dominate historically, but Japanese IP search interest in the US from 2020-2025 to be greater than US IPs over the same timeframe, which means that the US legacy commercial dominance does not align with present-day engagement.
