## Abstract

The United States has historically dominated the global entertainment market, with domestic intellectual properties (IPs) such as Marvel and Disney generating substantial global franchise revenue. However, in recent decades the global industry has experienced increasing competition from foreign IPs, particularly from Japan.  This project aims to evaluate the relationship between IP creation date, consumer engagement, and media exposure in order to determine which factor most strongly influences cumulative global revenue. It also examines how the influence of these factors differs between the U.S. and Japanese markets in terms of cumulative global revenue. The analysis was also performed separately for U.S. and Japanese IPs to compare differences between the two markets.
The results show that media exposure has the strongest relationship with cumulative global revenue overall, while consumer engagement measured by search interest also shows a positive relationship with revenue. Meanwhile, IP creation date shows no meaningful relationship with revenue. When comparing markets, Japanese IPs show a stronger relationship between search interest and revenue, while U.S. IPs show a stronger relationship between media exposure and revenue. These results show that modern popularity and media presence influence franchise revenue more than the age of the IP.

## Hypothesis


Generally, we think that consumer engagement is most important and will correlate the strongest with the cumulative global revenue of IPs, because engaged customers spend more money on an IP. In our project, an old creation date does not necessarily mean high engagement over time, and media exposure through airing platforms only gives a chance for fans to be engaged, while search interest reflects engagement more directly.

When we focus on US IPs, we expect media exposure to correlate the strongest with the global revenue of IPs for a simple reason——most US IPs are heavily dependent on TV/film adaptation when it comes to making money at a large scale.

When we focus on Japanese IPs, we expect IP creation date to correlate the strongest with the global revenue of Japanese IPs. We reason this because of the following: Japanese IPs are famous for their serialized mangas. Creating manga pairs high labor demand with a low budget for workers, which means that progressing the same amount in story would take longer through a manga than with film. This often causes an IP created through manga to stay long-running, providing more time to sell merchandize and their corresponding anime to run even longer than the manga itself, which allows Japanese IPs take a stronger advantage of peak popularity and have more chances at peak popularity if the IP cools down for some periods (lower dependence on a few peak consumer engagement) when it comes to cumulative global revenue. Manga, given that it is graphic, gains popularity fairly easily in the main entertainment markets (US, Japan, China, and Korea); note that Japan, China, and Korea all share similar aspects of culture through historical ties and their respective entertainment industries all have a very active manga/manhwa/webtoon community. This characteristic of the manga medium allows Japanese IPs to gain significant popularity even before being adapted into TV/film (lowering dependence on media exposure).

When comparing average search interest with cumulative global revenue, the Pearson r = 0.1990 indicates a weak positive linear relationship, and the Spearman ρ = 0.3456 indicates a moderate positive monotonic relationship. Both correlations are positive, meaning that IPs with higher search interest tend to have higher revenue. The Pearson p-value = 0.1530 is not statistically significant, and the Spearman p-value = 0.0113 is statistically significant. The fact that Spearman is stronger and significant suggests that revenue tends to increase with search interest in rank order, but the relationship is not always proportional. Search popularity is an indicator of IP success, but it only explains part of the revenue differences. When people frequently search for a franchise, it shows high fan engagement, active marketing campaigns, and new releases, which increase the likelihood that people will use or buy franchise products. 

For media exposure and cumulative global revenue, the Pearson r = 0.3349 shows a moderate positive linear relationship, while the Spearman ρ = 0.2356 shows a weak monotonic relationship. The positive correlations mean that IPs with more media exposure tend to generate more revenue. The Pearson p-value = 0.0163 is statistically significant, and the Spearman p-value = 0.0961 is not statistically significant, so media exposure appears to have some linear relationship with revenue. The weaker Spearman correlation suggests that the relationship may be influenced by outliers from the large franchises. Franchises that expand across movies, TV shows, games, merchandise, and books, etc. tend to accumulate greater revenue. There are more consumer entry points and different media that reach broader demographics. 

For the creation date and cumulative global revenue, the Pearson r = -0.0899 and Spearman ρ = -0.0886 are slightly negative, so older franchises might have a slightly higher revenue.
